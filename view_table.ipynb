```python
import pandas as pd
import matplotlib.pyplot as plt

# Leggi i CSV
uk = pd.read_csv('java_result/results_j_uk-2002.csv')
uk_transpose = pd.read_csv('java_result/results_j_uk-2002_transpose.csv')
webbase = pd.read_csv('java_result/results_j_webbase-2001.csv')
webbase_transpose = pd.read_csv('java_result/results_j_webbase-2001_transpose.csv')

# Unisci i dati in un unico DataFrame
data = pd.concat([uk, uk_transpose, webbase, webbase_transpose], keys=['.uk', 'Transpose', 'WebBase', 'Transpose'])

# Crea la tabella
fig, ax = plt.subplots(figsize=(10, 6))
ax.axis('off')
table = ax.table(cellText=data.values, colLabels=data.columns, loc='center')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carica il CSV
data = pd.read_csv('java_result/results_j_uk-2002.csv')

# Riorganizza i dati per ottenere le colonne con W=1, W=3, W=7
# Prima dividiamo i dati in base al CompressionWindow (W)
df_w1 = data[data['CompressionWindow'] == 1].reset_index(drop=True)
df_w3 = data[data['CompressionWindow'] == 3].reset_index(drop=True)
df_w7 = data[data['CompressionWindow'] == 7].reset_index(drop=True)

# Crea una nuova tabella combinando i dati per le colonne W=1, W=3, W=7
table_data = {
    'R': df_w1['MaxRefCount'].tolist(),
    'W=1 Avg Ref Chain': df_w1['AvgRefChain'].tolist(),
    'W=3 Avg Ref Chain': df_w3['AvgRefChain'].tolist(),
    'W=7 Avg Ref Chain': df_w7['AvgRefChain'].tolist(),
    'W=1 Bits/Node': df_w1['BitsNode'].tolist(),
    'W=3 Bits/Node': df_w3['BitsNode'].tolist(),
    'W=7 Bits/Node': df_w7['BitsNode'].tolist(),
    'W=1 Bits/Link': df_w1['BitsLink'].tolist(),
    'W=3 Bits/Link': df_w3['BitsLink'].tolist(),
    'W=7 Bits/Link': df_w7['BitsLink'].tolist(),
}

df_table = pd.DataFrame(table_data)

# Impostare la dimensione della figura
fig, ax = plt.subplots(figsize=(12, 6))

# Nascondere gli assi
ax.axis('tight')
ax.axis('off')

# Creazione della tabella
table = ax.table(cellText=df_table.values, 
                colLabels=df_table.columns, 
                loc='center', 
                cellLoc='center', 
                colColours=['#f1f1f1']*df_table.shape[1])

# Ottenere la cella
cell = table.get_celld()

# Unire le celle per "Average Reference Chain", "Bits/Node", e "Bits/Link"
cell[(0, 1)].set_text_props(text="Avg Ref Chain")
cell[(0, 2)].set_text_props(text="")
cell[(0, 3)].set_text_props(text="")
cell[(0, 4)].set_text_props(text="Bits/Node")
cell[(0, 5)].set_text_props(text="")
cell[(0, 6)].set_text_props(text="")
cell[(0, 7)].set_text_props(text="Bits/Link")
cell[(0, 8)].set_text_props(text="")
cell[(0, 9)].set_text_props(text="")

# Aggiungere la seconda riga con le intestazioni "W=1", "W=3", "W=7"
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

# Impostare la seconda riga con W=1, W=3, W=7
cell[(1, 1)].set_text_props(text="W=1")
cell[(1, 2)].set_text_props(text="W=3")
cell[(1, 3)].set_text_props(text="W=7")
cell[(1, 4)].set_text_props(text="W=1")
cell[(1, 5)].set_text_props(text="W=3")
cell[(1, 6)].set_text_props(text="W=7")
cell[(1, 7)].set_text_props(text="W=1")
cell[(1, 8)].set_text_props(text="W=3")
cell[(1, 9)].set_text_props(text="W=7")

# Mostra la tabella
plt.show()
```
